```python

columns_to_clean = ['trtbps','chol', 'thalachh', 'oldpeak']

for column in columns_to_clean:
    # Calculate Q1, Q3, and IQR for the current column
    Q1 = heart_df[column].quantile(0.25)
    Q3 = heart_df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outlier removal
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers from the current column
    heart_df[column] = np.where((heart_df[column] < lower_bound) | (heart_df[column] > upper_bound),
                                 np.where(heart_df[column] < lower_bound, lower_bound, upper_bound),
                                 heart_df[column])
```
